```python
import numpy as np
```

```python
matrix = np.random.randint(1, 90, (5, 5))
print("Matrix: ", matrix)

row_mean = np.mean(matrix, axis=1, dtype=int)
print("Row_mean", row_mean)

col_mean = np.mean(matrix, axis=0, dtype=int)
print("Col_mean: ", col_mean)

print("Sorted_row: ", np.sort(matrix, axis=1))
print("Sorted_col: ", np.sort(matrix, axis=0))

row_sum = np.sum(matrix, axis=1)
col_sum = np.sum(matrix, axis=0)
print("Sum of Row: ", row_sum)
print("Sum of Col: ", col_sum)

matrix = np.where(matrix % 2 == 0, -1, matrix)
print(matrix)

neg_count = np.sum(matrix < 0, axis=0)
print(neg_count)
```

```python
mat_stock = np.random.randint(20, 80, (5, 5))
print("Stock: ", mat_stock)

mat_demand = np.random.randint(20, 80, (5, 5))
print("Demand: ", mat_demand)

remaining = mat_stock - mat_demand
print(remaining)

total_shortage = np.sum(np.where(remaining < 0, -remaining, 0))
print(total_shortage)

neg_values = np.where(remaining < 0, 0, remaining)
print(neg_values)

zero_stock = np.sum(np.where(remaining > 1, 1, remaining))
print(zero_stock)

print("Sorted_row: ", np.sort(remaining, axis=1))
```
