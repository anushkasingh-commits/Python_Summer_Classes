```python
import pandas as pd
```

```python
df = pd.read_csv('student_data.csv')
```

```python
df.groupby('placement_status')['placement_status'].agg('count')
```

```python
df.groupby('placement_status')['exam_score'].agg('mean')
```

```python
df.groupby('placement_status')['study_hours'].agg('mean')
```

```python
df.groupby('exam_score')['study_hours'].agg('mean')
```

```python
def grades(x):
  if x >= 90:
    return "A"
  elif x >= 80:
    return "B"
  elif x >= 70:
    return "C"
  else:
    return "D"
```

```python
df['grades']=df['exam_score'].apply(grades)
df
```
